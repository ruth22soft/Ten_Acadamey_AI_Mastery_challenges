```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#Load the dataset
data1 = pd.read_csv('D:/Git_clones/Ten_Acadamey_AI_Mastery_challenges/data/benin-malanville.csv')
data2 = pd.read_csv('D:/Git_clones/Ten_Acadamey_AI_Mastery_challenges/data/sierraleone-bumbuna.csv')
data3 = pd.read_csv('D:/Git_clones/Ten_Acadamey_AI_Mastery_challenges/data/togo-dapaong_qc.csv')
```

```python
print(data1.describe())
print(data2.describe())
print(data3.describe())
# Data Quality Check
# Check for missing values
print(data1.isnull().sum())
print(data2.isnull().sum())
print(data3.isnull().sum())
```

```python
for col in ['GHI', 'DNI', 'DHI']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data1[col])
    plt.title(f'Boxplot of benin{col}')
    plt.show()

for col in ['GHI', 'DNI', 'DHI']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data2[col])
    plt.title(f'Boxplot of sierraleone{col}')
    plt.show()

for col in ['GHI', 'DNI', 'DHI']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data3[col])
    plt.title(f'Boxplot of togo{col}')
    plt.show()

```

```python
# Time Series Analysis:
# Plot GHI, DNI, DHI, and Tamb over time
plt.figure(figsize=(12, 6))
plt.plot(data1['Timestamp'], data1['GHI'], label='GHI')
plt.plot(data1['Timestamp'], data1['DNI'], label='DNI')
plt.plot(data1['Timestamp'], data1['DHI'], label='DHI')
plt.plot(data1['Timestamp'], data1['Tamb'], label='Tamb')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.legend()
plt.title('Time Series of Solar Radiation and Temperature')
plt.show()
```
